```python
#Problem 1
import numpy as np
np.random.seed(0)
mean = [-3, 0]
cov = [[1.0, 0.8], [0.8, 1.0]]
size = [500]
A_d = np.random.multivariate_normal(mean, cov, size)
#A_d.size
A_d.shape
#print(A_d)
```
