```python
# Problem 5
def compute_gradient(function, x_range=(-50, 0, 50.1, 0.1)):

  X = np.arange(x_range[0],x_range[2],x_range[3])
  X3 = np.arange(x_range[1],x_range[2],x_range[3])
  y = function(X)
  y3 = function(X3)

  change_X = X[1:] - X[:-1]
  change_y = y[1:] - y[:-1]
  change_X3 = X3[1:] - X3[:-1]
  change_y3 = y3[1:] - y3[:-1]

  gradient = change_y/change_X
  gradient3 = change_y3/change_X3

  pass
  if function == function3:
    return X3, y3, gradient3
  else:
    return X, y, gradient
  

def function1(array_X):
  array_y = array_X ** 2
  pass
  return array_y
def function2(array_X):
  array_y2 = array_X*2 +(2*(array_X**2))
  pass
  return array_y2 
def function3(array_X):
  array_y3 = np.sin((array_X)*0.5)
  pass
  return array_y3

x1, y1, gradient1 = compute_gradient(function1)
x2, y2, gradient2 = compute_gradient(function2)
x3, y3, gradient3 = compute_gradient(function3)
```
