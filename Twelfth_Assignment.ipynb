```python
# Problem 7(b)
def corrFilter(x: pd.DataFrame, bound: float):
    xCorr = x.corr()
    xFiltered = xCorr[((xCorr >= bound) | (xCorr <= -bound)) & (xCorr !=1.000)]
    xFlattened = xFiltered.unstack().sort_values().drop_duplicates()
    return xFlattened

corrFilter(df, .641)
```

```python
# Problem 7(c)
dfCorr = df.corr()
filteredDf = dfCorr[((dfCorr >= .641) | (dfCorr <= -.641)) & (dfCorr !=1.000)]
plt.figure(figsize=(30,10))
sns.heatmap(filteredDf, annot=True, cmap="Reds")
plt.show()
```

```python
# Problem 7(e)
def corrFilter(x: pd.DataFrame, bound: float):
    xCorr = x.corr()
    xFiltered = xCorr[((xCorr >= bound) | (xCorr <= -bound)) & (xCorr !=1.000)]
    xFlattened = xFiltered.unstack().sort_values().drop_duplicates()
    return xFlattened

corrFilter(df, .82)
```

```python
# Problem 7(f)
dfCorr = df.corr()
filteredDf = dfCorr[((dfCorr >= .82) | (dfCorr <= -.82)) & (dfCorr !=1.000)]
plt.figure(figsize=(30,10))
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
sns.set_style('white')
sns.heatmap(filteredDf, annot=True, cmap="Reds", annot_kws={'size':16})
plt.show()
```
